```python
import matplotlib.pyplot as plt
import numpy as np

# ================================
# 1. BAR CHARTS
# ================================

# Sample data for bar charts (e.g., Monthly Sales)
categories_bar = ["January", "February", "March", "April", "May", "June"]
values_bar = [100, 150, 130, 170, 160, 180]
colors = ["red", "blue", "green", "purple", "orange", "cyan"]

# --- Bar Chart with Data Point Labels ---
plt.figure(figsize=(10, 6))
bars = plt.bar(categories_bar, values_bar, color = colors, label="Sales")
# Adding data point labels on top of each bar
for i, value in enumerate(values_bar):
    plt.text(i, value + 2, str(value), ha='center', va='bottom', fontsize=10)

plt.xlabel("Month")
plt.ylabel("Sales (in thousands)")
plt.title("Monthly Sales - Bar Chart with Data Labels")
plt.xticks(rotation=45)
plt.legend(title="Legend")
plt.grid(True)
plt.savefig(f"bar_chart_test_1.png")
plt.show()

# --- Bar Chart without Data Point Labels ---
plt.figure(figsize=(10, 6))
plt.bar(categories_bar, values_bar, color = colors, label="Sales")
plt.xlabel("Month")
plt.ylabel("Sales (in thousands)")
plt.title("Monthly Sales - Bar Chart without Data Labels")
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig(f"bar_chart_test_2.png")
plt.show()
```

```python
# ================================
# 2. LINE CHARTS
# ================================

# Sample data for line charts (e.g., Yearly Revenue)
years = np.array([2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022])
revenue = np.array([400, 450, 480, 520, 490, 500, 600, 550, 650, 700])


# --- Line Chart with Data Point Labels ---
plt.figure(figsize = (15, 8))
plt.plot(years, revenue, marker='o', linestyle='-', color='green', label="Revenue")

# Adding data point labels near each marker
for x, y in zip(years, revenue):
    plt.text(x, y + 5, f"{y}", ha='center', va='bottom', fontsize=10)

plt.xlabel("Year")
plt.ylabel("Revenue (in millions)")
plt.title("Yearly Revenue - Line Chart with Data Labels")

# Set alternating X-axis ticks
plt.xticks(years[::2])  # Show every other year
plt.grid(True)
plt.savefig("line_chart_test_1.png")
plt.show()


# --- Line Chart without Data Point Labels ---
plt.figure(figsize=(15, 8))
plt.plot(years, revenue, marker='o', linestyle='-', color='purple', label="Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue (in millions)")
plt.title("Yearly Revenue - Line Chart without Data Labels")
plt.xticks(years)
plt.grid(True)
plt.savefig(f"line_chart_test_2.png")
plt.show()
```

```python
# ================================
# 3. PIE CHARTS
# ================================

import matplotlib.pyplot as plt

# Sample data for pie charts (e.g., Market Share by Region)
regions = ["North America", "Europe", "Asia", "South America", "Africa"]
market_share = [40, 20, 25, 10, 5]

# --- Pie Chart with Data Point Labels ---
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(market_share, labels=regions, autopct='%1.1f%%',
                                   startangle=90, shadow=False)
plt.title("Market Share by Region - Pie Chart with Data Labels")
plt.legend(wedges, regions, title="Regions", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.savefig("pie_chart_test_1.png", bbox_inches='tight')
plt.show()

# --- Pie Chart without Data Point Labels ---
plt.figure(figsize=(8, 8))
wedges, texts = plt.pie(market_share, labels=regions, startangle=90, shadow=False)
plt.title("Market Share by Region - Pie Chart without Data Labels")
plt.legend(wedges, regions, title="Regions", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.savefig("pie_chart_test_2.png", bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for pie charts (e.g., Market Share by Region)
regions = ["North America", "Europe", "Asia", "South America", "Africa"]
market_share = [40, 20, 25, 10, 5]

# --- Pie Chart with Inner Percentage Numbers Only (No outer labels) ---
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(market_share, labels=None, autopct='%1.1f%%',
                                   startangle=90, shadow=False)
plt.title("Market Share by Region - Inner Percentage Numbers Only")
plt.legend(wedges, regions, title="Regions", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.savefig("pie_chart_test_3.png", bbox_inches='tight')
plt.show()

# --- Pie Chart Without Percentage Numbers and Outer Labels ---
plt.figure(figsize=(8, 8))
wedges, texts = plt.pie(market_share, labels=None, autopct=None,
                        startangle=90, shadow=False)
plt.title("Market Share by Region - Plain Pie Chart")
plt.legend(wedges, regions, title="Regions", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.savefig("pie_chart_test_4.png", bbox_inches='tight')
plt.show()
```
